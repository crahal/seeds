```python
import random
import seaborn as sns

def coinToss(number):
    head_counter = 0
    for i in range(number):
        if random.randint(0, 1)==0: head_counter += 1
    return head_counter

for tosses in [100, 1000, 5000]:
    meta_counter = []
    for seed in range(10000):
        random.seed(seed)
        meta_counter.append(coinToss(tosses)/tosses)
    sns.kdeplot(meta_counter, common_grid=True)
```
